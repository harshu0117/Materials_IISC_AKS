# Loading the original base model ( LLAMA 3 8B ) via unsloth 

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

# Load your model from Hugging Face
model_name = "unsloth/llama-3-8b-bnb-4bit"
tokenizer = AutoTokenizer.from_pretrained(model_name)
model_1 = AutoModelForCausalLM.from_pretrained(model_name, torch_dtype=torch.float16, device_map="auto")

print("Model 1 loaded successfully")
```

## Generating responses from base model

```python

prompts = [
    "List the best thermoelectric materials today and compare the top five in terms of Seebeck coefficient, electrical conductivity, and thermal conductivity.",
    "Explain why these thermoelectric materials are considered the best—what individual properties enhance their performance and quantify their contributions.",
    "What makes different thermoelectric materials exhibit varying thermoelectric behavior? Discuss how materials chemistry relates to their thermal and electronic properties.",
    "Is it possible to discover a better thermoelectric material? If so, what improvements or novel strategies should be pursued to break current limitations?",
    "How do thermal and electronic properties affect thermoelectric efficiency? Can these dependencies be decoupled, and how might that be achieved in advanced materials?"
]


gen_args = {
    "max_new_tokens": 200,
    "do_sample": False,      # Use greedy decoding—highest-probability token every step
    "num_beams": 1,          # Single beam → vanilla greedy search
    "temperature": None,     # Ignored when sampling is off
    "top_k": None,           # Ignored
    "top_p": None,           # Ignored
    "repetition_penalty": 1.0
}



# Run each prompt
for idx, prompt in enumerate(prompts, 1):
    inputs = tokenizer(prompt, return_tensors="pt").to("cuda")
    outputs = model_1.generate(
        **inputs,
        pad_token_id=tokenizer.pad_token_id,
        **gen_args
    )
    response = tokenizer.decode(outputs[0], skip_special_tokens=True)
    print(f"\nPrompt {idx}: {prompt}\nResponse {idx}: {response.strip()}")
```
